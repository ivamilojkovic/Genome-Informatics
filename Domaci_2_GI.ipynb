```python
import pysam

bam_file = pysam.AlignmentFile("merged-tumor.bam", "rb")  
```

```python
number_unmapped = 0
quality_zero = 0
total_quality = 0
total = 0

for read in bam_file:
  if read.is_unmapped:
    number_unmapped += 1

  if read.mapping_quality==0:
    quality_zero += 1
    
  total_quality += read.mapping_quality
  total += 1

  if total==1:
    first_read = read
    print("The first read from the AlignmentFile:")
    print(first_read, "\n")

avg_all = total_quality/total
avg_not_all = total_quality/(total-quality_zero)

print("Unmapped reads: ", number_unmapped)
print("Total number of reads: ", total)
print("Number of reads with mapping quality 0: ", quality_zero)
print("Average mapping quality for all the reads: ", avg_all)
print("Average mapping quality without reads with 0 mapp quality: ", avg_not_all, "\n")
```

The first read from the BAM file represents an AlignedSegment and it containes the following fields:

*   QNAME 
*   FLAG
*   RNAME
*   POS
*   MAQ
*   CIGAR...

Specifically, the FLAG field respresents some additional information about the read. Here this field has value 1187 which is in binary representation 10010100011. The encoded properties of the read are:

* Read is paired
* Read is mapped in proper pair
* Mate reverse strand
* Second in pair
* Read is PCR or optical duplicate
